```
import numpy as np
from numpy.random import randn
from scipy import stats

import seaborn as sns

%matplotlib inline
```

```
data1 = randn(100)
data2 = randn(100) + 2
```

```
sns.distplot(data1)
sns.distplot(data2)
```

```
sns.boxplot(data=[data1, data2]) # ボックスプロットを描画する
```

```
sns.boxplot(data=[data1,data2], whis=np.inf) # 髭の範囲を無限大にすると外れ値を含むことができる
```

```
sns.boxplot(data=[data1, data2], orient='h') # グラフの向きを横にする
```

```
data1 = stats.norm(0,5).rvs(100)
```

```
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, -1*stats.gamma(5).rvs(50)])
```

```
sns.boxplot(data=[data1, data2])
```

```
sns.violinplot(data=[data1, data2]) # バイオリンプロットだとボックスプロットでは分からない分布を見ることができる
```

```
sns.violinplot(data=data2, bw=0.01) # バイオリンプロットはカーネル密度推定を使用して算出されているのでバンド幅を変えることができる
```

```
sns.violinplot(data=data1, inner='stick') # バイオリンプロットないにラグプロットを出力できる
```
