```
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
```

```
%matplotlib inline
```

```
dataset1 = randn(100)
```

```
plt.hist(dataset1) # ヒストグラムを表示する
```

```
dataset2 = randn(80)
```

```
plt.hist(dataset2, color='indianred') # 色を指定してヒストグラムを表示
```

```
plt.hist(dataset1, density=True, bins=20) # 縦軸を標準化してヒストグラムを表示
```

```
plt.hist(dataset1, density=True, alpha=0.5, bins=20)
plt.hist(dataset2, density=True, alpha=0.5, bins=20, color='indianred')
```

```
data1 = randn(1000)
data2 = randn(1000)
```

```
sns.jointplot(data1, data2) # 同時分布、結合分布
```

```
sns.jointplot(data1, data2, kind='hex') # スキャッタープロットを六角形で二次元のヒストグラムのように表示
```
