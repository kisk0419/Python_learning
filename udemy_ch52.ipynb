```
import numpy as np
from numpy.random import randn

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```
tips = sns.load_dataset('tips') # Seabornのテストデータを取得する
```

```
tips
```

```
sns.lmplot('total_bill', 'tip', tips) # X軸にtotal_bill、Y軸にtipをとって回帰直線とスキャッタープロットを描画
```

```
sns.lmplot('total_bill', 'tip', tips, 
           scatter_kws={'marker':'o', 'color':'indianred'},
           line_kws={'linewidth':1, 'color':'blue'})
```

```
sns.lmplot('total_bill', 'tip', tips, order=4, # 4次関数で近似する
           scatter_kws={'marker':'o', 'color':'indianred'},
           line_kws={'linewidth':1, 'color':'blue'})
```

```
sns.lmplot('total_bill', 'tip', tips, fit_reg=False) # 回帰直線をかかない
```

```
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])
```

```
tips.tip_pect
```

```
sns.lmplot('size', 'tip_pect', tips) # 人数のように離散的な値でも描画できる
```

```
sns.lmplot('size', 'tip_pect', tips, x_jitter=0.2) # スキャッタープロットの表示幅を拡張する
```

```
sns.lmplot('size', 'tip_pect', tips, x_estimator=np.mean) 
```

```
sns.lmplot('total_bill', 'tip_pect', tips, hue='sex', markers=['x','o']) # hueで指定したカラムの値を種別で出し分けてくれる
```

```
sns.lmplot('total_bill', 'tip_pect', tips)
```

```
sns.lmplot('total_bill', 'tip_pect', tips, hue='day')
```

```
sns.lmplot('total_bill', 'tip_pect', tips, lowess=True, line_kws={'color':'black'}) # 回帰直線をなめらかにする
```

```
sns.regplot('total_bill', 'tip_pect', tips)
```

```
fig, (axis1, axis2) = plt.subplots(1, 2, sharey=True) # チャートの表示領域を並べることができる
sns.regplot('total_bill', 'tip_pect', tips, ax=axis1)
sns.violinplot(y='tip_pect', x='size', data=tips.sort_values('size'), ax=axis2)
```
