```
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from urllib.request import urlretrieve
```

```
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
```

```
savepath='winequality-red.csv'
urlretrieve(url, savepath)
```

```
!head -n 10 winequality-red.csv
```

```
dframe_wine = pd.read_csv('winequality-red.csv', sep=';')
```

```
dframe_wine.head()
```

```
dframe_wine['alcohol'].mean()
```

```
def max_to_min(arr):
  return arr.max() - arr.min()
```

```
wino = dframe_wine.groupby('quality')
```

```
wino.describe()
```

```
wino.agg(max_to_min) # 指定した関数を使用して集計
```

```
wino.agg('mean')
```

```
dframe_wine['qua/alc ration'] = dframe_wine['quality'] / dframe_wine['alcohol'] # 列の追加
```
